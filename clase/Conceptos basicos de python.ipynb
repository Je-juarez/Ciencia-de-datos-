ejericio, crear una funcon que tome como valores de entrada los coeficientes de una ecuacion cuadratica y devuelva las raices. 

```python

from math import sqrt, exp
def fgral(a, b, c):

    d= ((b)**2)-(4*c*a)
    if d >=0:
      x1= ((-b)+ (sqrt(((b)**2)-(4*c*a))))/(2*a)
      x2= ((-b)-(sqrt(((b)**2)-(4*c*a))))/(2*a)
      return x1, x2
    else: print("no tiene solución")
```
