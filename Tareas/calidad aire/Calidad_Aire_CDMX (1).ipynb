```python
#Convertir objetos a flotantes de las delegaciones de interes.
SMCA['CUA'] = pd.to_numeric(SMCA['CUA'],errors = 'coerce')
SMCA['MGH'] = pd.to_numeric(SMCA['MGH'],errors = 'coerce')
SMCA['IZT'] = pd.to_numeric(SMCA['IZT'],errors = 'coerce')
SMCA['MPA'] = pd.to_numeric(SMCA['MPA'],errors = 'coerce')
print(SMCA.info())

```

```python
#Dejar solo las delegaciones de interes
SMCA_1=SMCA.drop(['ACO', 'AJM', 'ATI', 'BJU', 'CAM','CCA', 'CHO', 'FAC', 'HGM', 'INN', 'LLA', 'LPR', 'MER', 'MON', 'NEZ', 'PED', 'SAG', 'SFE', 'TAH', 'TLA', 'TLI','UAX', 'UIZ', 'VIF', 'XAL', 'TAH.1', 'TLA.1', 'TLI.1', 'TPN', 'UAX.1','UIZ.1', 'VIF.1', 'XAL.1'], axis=1)
SMCA_1
```

```python
CO= SMCA_1 ["CONTAMINANTE"]== "CO"
SMCA_CO = SMCA_1[CO]
SMCA_CO.head()
```

```python
#Quitar la hora
SMCA_CO_DIA=SMCA_CO.drop(["HORA"], axis=1)
#Promedio por día
SMCA_CO_DIA.groupby("Fecha").mean()

```

Separar el DF por contaminacion al dia para poder comparar y establecer su nivel de riesgo.

```python
NO2= SMCA_1 ["CONTAMINANTE"]== 'NO2'
SMCA_NO2 = SMCA_1[NO2]
SMCA_NO2.head()
```

```python
#Quitar la hora
SMCA_NO2_DIA=SMCA_NO2.drop(["HORA"], axis=1)
#Promedio por día
SMCA_NO2_DIA = SMCA_NO2_DIA.groupby("Fecha").mean()
#grafica
graf_NO2 = ["CUA", "IZT", "MGH", "MPA"]
ejes = SMCA_NO2_DIA[graf_NO2].plot(marker='.', alpha=0.1, linestyle='None',figsize=(10,10),subplots=True)
for eje in ejes:
  eje.set_ylabel('Niveles de NO2 en ppm')
```
